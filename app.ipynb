```python
from flask import Flask, render_template, request, redirect, url_for
import os
# Import your logic from your specific file
from recommender import get_skincare_roadmap 

app = Flask(__name__)

# Configuration for where to save uploaded images
app.config['UPLOAD_FOLDER'] = 'static/uploads'

# Create the uploads folder automatically if it doesn't exist
os.makedirs(app.config['UPLOAD_FOLDER'], exist_ok=True)

@app.route('/')
def home():
    """Renders the landing page (index.html)"""
    return render_template('index.html')

@app.route('/analyze', methods=['POST'])
def analyze():
    """Handles image upload and triggers the AI roadmap"""
    if 'file' not in request.files:
        return redirect(request.url)
    
    file = request.files['file']
    
    if file.filename == '':
        return redirect(request.url)

    if file:
        # Save the file so our model can access it
        filepath = os.path.join(app.config['UPLOAD_FOLDER'], file.filename)
        file.save(filepath)
        
        # 🚀 CALL YOUR RECOMMENDER LOGIC
        # This returns the dictionary we built earlier
        analysis_result = get_skincare_roadmap(filepath)
        
        # If confidence was too low (<65%), analysis_result will be None
        if analysis_result is None:
            return render_template('index.html', error="AI Analysis Inconclusive. Please use a clearer photo.")

        # Pass the dictionary and the image filename to the results page
        return render_template('result.html', result=analysis_result, img_file=file.filename)

if __name__ == '__main__':
    # Start the local server
    app.run(debug=True)
```
